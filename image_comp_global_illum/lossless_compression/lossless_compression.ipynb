```python
import math

def sign_entropy(sign_probability):
    return -math.log(sign_probability, 2)

def sign_mean_entropy(sign_probability):
    return sign_probability * sign_entropy(sign_probability)
```

```python
import pprint

input_ = 'FRISCHE_FISCHE_ZISCHEN'
dict_ = {}

# Calculate the propability for each sign
for i in input_:
    if i in dict_:
        dict_[i]['count'] += 1
        
    else:
        dict_[i] = {}
        dict_[i]['count'] = 1
        
    dict_[i]['probab'] = dict_[i]['count'] / float(len(input_))
        
sum_ = 0
I_tot = 0

for k in dict_:
    dict_[k]['I'] = sign_entropy(dict_[k]['probab'])
    dict_[k]['H'] = sign_mean_entropy(dict_[k]['probab'])
    
    sum_ += dict_[k]['H']

I_tot = sum_ * len(input_)
        
print pprint.pprint(dict_)

print ''

print 'Sum (H): {0}'.format(sum_)
print 'I_tot: {0}'.format(I_tot)

print ''

bit_orig = len(input_) * 8
print 'Bit original: {0} Bit'.format(bit_orig)
print 'Redundancy: {0} bit'.format(bit_orig - I_tot)
```
